```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 1)
y1 = 10**(-10*(x-0.1))
y2 = 4**(-4*(x-0.1))
y3 = 1-((x-0.1)/0.9)
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,6))
ax.plot(x,y1, color = 'midnightblue', linewidth = 2, label = 'TP1: $10^{-10(t_i-0.1)}$')
ax.plot(x,y2, color = 'dodgerblue', linewidth = 2, label = 'TP2: $4^{-4(t_i-0.1)}$')
ax.plot(x,y3, color = 'lightskyblue', linewidth = 2, label = 'TP3: Linear decrease')
ax.plot([0.1,0.1], [-0.1,1.1], 'k--', linewidth = 1)
ax.plot([0,0.1], [1,1], 'k', linewidth=4, alpha = 0.6)
ax.plot([1,1.1], [0.003,0.003], 'k', linewidth=4, alpha = 0.6)
plt.legend(fontsize = 16)
ax.set_ylim([-0.02,1.02])
ax.set_xlim([0,1.1])
ax.tick_params(which='both', labelsize = 16)
ax.set_xlabel('Ice lens thickness (m)', fontsize = 20)
ax.set_ylabel('Permeability', fontsize = 20)
plt.savefig('ice_lens_models.png', dpi=300, bbox_inches = 'tight', transparent = True)
```
